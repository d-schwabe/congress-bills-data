## 5.5: Using Altair to visualize key indicators of word count trends 

#### Graphing the mean senate bill word count by congress

```python
bar = alt.Chart(summary).mark_bar().encode(
    x='congress',
    y='mean'
)
rule = alt.Chart(summary).mark_rule(color='red').encode(
    y='mean(mean):Q'
)

(bar + rule).properties(
    title='Average senate bill wordcount over 13 Congresses, 1995-2020'
)
```
